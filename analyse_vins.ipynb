```python
#Import des bibliothèques
import pandas as pd
import seaborn as sns

link = "https://github.com/WildCodeSchool/wilddata/raw/main/wine.zip"
df = pd.read_csv(link)

df.head()
```

```python
#Affichage des dimenssions du DataFrame (nombres de lignes et de colonnes)
df.shape
```

```python
#Fonction pour afficher les informations principales du dataFrame
df.info()
```

```python
#Fonction qui affiche les statistiques des valeurs numériques
df.describe()
```

```python
#Arrondir les valeurs des colonnes 'points' et 'price' à deux chiffres après la virgule
df[['points', 'price']] = df[['points', 'price']].round(2)
```

```python
#Calcul du pourcentage de valeurs nulles par colonnes
df.isnull().mean() * 100
```

```python
#Supprimer les lignes avec des valeurs nulles dans la colonne 'price'
df.dropna(subset=['price'], inplace=True)
```

```python
#Affichage des valeurs dupliquées
df.duplicated()
df[df.duplicated()]
```

```python
#Fonction pour supprimer les valeurs dupliquées
df.drop_duplicates(inplace=True)
```

```python
#Affichage des valeurs uniques pour la colonne 'country'
df.country.unique()
```

```python
#Affichage des valeurs uniques pour la colonne 'variety'
df.variety.unique()
```

```python
#Affichage des valeurs uniques pour la colonne 'variety'
df.points.unique()
```

```python
#Verification des valeurs uniques de la colonne 'millesime"
df['millesime'].unique()
```

```python
#Filtrer les valeurs de 'millesime' entre 1800 et 2024 pour enlever les valeurs abérantes
df = df.loc[(df['millesime'] >= 1800) & (df['millesime'] <= 2024)]

#Verification des valeurs uniques de la colonne 'millesime"
df['millesime'].unique()
```

# Visualisation

```python
#Import de la bibliotheque matplotlib pour la visualisation
import matplotlib.pyplot as plt

#Histogramme des points
plt.hist(df['points'], bins=20, edgecolor='black')
plt.xlabel('Points')
plt.ylabel('Fréquence')
plt.title('Distribution des points')
plt.show()
```

```python
#Diagramme circulaire des 5 pays les plus présents dans le DataFrame

# Calcul des fréquences des pays
calcul_top_pays = df['country'].value_counts()

#Sélection des 5 premiers pays
top_countries = calcul_top_pays.head(5)

# Tracage le diagramme circulaire
plt.figure(figsize=(6, 6))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 des pays représentés')

plt.show()
```

```python
#Horizontal barplot du top 10 des variétés de vins les plus représentés

#Calculer le nombre d'occurrences de chaque variété de vin et son top 10
top_vin = df['variety'].value_counts().head(10)

#Trier les données par ordre croissant
top_vin = top_vin.sort_values(ascending=True)

#Tracer le graphique à barres horizontal pour les top 10 variétés de vin
plt.figure(figsize=(6, 5))
plt.barh(top_vin.index, top_vin.values, color='skyblue')
plt.ylabel('Variété de vin')
plt.xlabel('Nombre de vins')
plt.title('Top 10 des variétés de vin')
plt.tight_layout()
plt.show()
```

```python
#Création d'une heatmap pour visualiser les corrélations
corr_matrix = df[['points', 'price', 'millesime']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Corrélation entre les points, les prix et les millesime')
plt.show()
```
